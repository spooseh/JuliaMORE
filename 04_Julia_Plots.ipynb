# Julia Plots

There are many "[Graphics](https://juliapackages.com/c/graphics)" packages.
- [Plots.jl](https://docs.juliaplots.org/latest/) (1267⭐)
    * metapackage wrapping several backends like GR, Plotly, PythonPlot,
        - switch backends: gr() and later plotly() 
    * [StatsPlots.jl](https://github.com/JuliaPlots/StatsPlots.jl) (256⭐)
        * a replacement of Plots.jl with many statistical recipes   
        * works directly on DataFrames    
- [PyPlot.jl](https://makie.juliaplots.org/stable/) (357⭐)
    * latest release Mar 2023
- [Makie.jl](https://makie.juliaplots.org/stable/) (885⭐)
    * latest release Jan 2023
- [Gadfly.jl](http://gadflyjl.org/stable/) (1659⭐)
    * latest release Oct 2021, like ggplot

## Plot random walks

```julia
using Plots
gr() # default, to use PyPlot or others you need to install the package first and switch the backend 
"""
    rnd_walk(init, n_walker, n_step)
Generate a matrix of random walks from a common initial state
"""
function rnd_walk(init, n_walker, n_step)
    ini_states = fill(init, 1, n_walker) # vector of initial states
    rnd_states = rand([-1,1], n_step, n_walker) # matrix of random states, each column a walk
    return cumsum(vcat(ini_states, rnd_states), dims=1) # cumulative sum over columns, return optional
end

using Random # use StableRNG for reproduciblity among different Julia versions
Random.seed!(1984)

init = 0 # initial state
n_walker = 10
n_step = 100
walks = rnd_walk(init, n_walker, n_step);
```

```julia
p1 = plot(walks, legend=nothing)
```

```julia
plot!(walks[:,end], lw=3, color=:red)
```

```julia
p2 = histogram(walks[:], bins=20, legend=nothing)
```

```julia
plot(p1, p2, layout=(1,2))
```
